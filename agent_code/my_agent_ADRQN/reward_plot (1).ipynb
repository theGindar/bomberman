```python
import matplotlib.pyplot as plt
import pickle as pkl
import numpy as np
```

```python
with open('./rewards/22 rewards 8460 (mb, mrs, g0,9, random_threshold).pkl', 'rb') as f:
    rewards_random = pkl.load(f)
with open('./rewards/21 rewards 8460 (mb, mrs, g0.9, eps_threshold).pkl', 'rb') as f:
    rewards_eps = pkl.load(f)
```

```python
rewards_random = np.convolve(rewards_random, np.ones(70,)/70, mode='same')
rewards_random = rewards_random[70:len(rewards_random)-70]

rewards_eps = np.convolve(rewards_eps, np.ones(70,)/70, mode='same')
rewards_eps = rewards_eps[70:len(rewards_eps)-70]


coef = np.polyfit(range(len(rewards_random)), rewards_random, 1)
poly1d_fn = np.poly1d(coef)
x = range(len(rewards_eps))
```

```python

#plt.plot(x,rewards, x, poly1d_fn(x), '--k')
fig, ax0 = plt.subplots()
ax0.plot(rewards_random[:(len(rewards_eps)-1)], color="blue", label="Constant random threshold")
ax0.plot(rewards_eps[:(len(rewards_eps)-1)], color="red", label="Epsilon greedy threshold")
plt.ylabel("Reward")
plt.xlabel("Episodes")
plt.legend()
# plt.title("Total Rewards per Round")
plt.savefig("rewards_eps_vs_random.pdf")
plt.show()
```
